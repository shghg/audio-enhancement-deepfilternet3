"""
DeepFilterNet Audio Enhancement Script

This script applies noise suppression to a given noisy audio file using DeepFilterNet,
an advanced deep-learning-based speech enhancement model. The script follows these steps:

1. **Initialize the DeepFilterNet Model**:
   - Loads a pre-trained DeepFilterNet model.
   - Sets up the necessary processing state (`df_state`).

2. **Load Noisy Audio**:
   - Reads the input noisy audio file.
   - Ensures the sample rate (`sr`) is retained.

3. **Convert Audio to Tensor Format**:
   - Converts the NumPy array representation of the audio into a PyTorch tensor.
   - Adds a batch dimension to the tensor, as required by DeepFilterNet.

4. **Enhance Audio Using DeepFilterNet**:
   - Passes the noisy audio through the pre-trained model.
   - Receives a cleaned/enhanced version of the audio.

5. **Convert Enhanced Audio Back to NumPy**:
   - Removes the batch dimension from the output tensor.
   - Converts the PyTorch tensor back to a NumPy array.

6. **Save the Enhanced Audio**:
   - Writes the processed audio to an output `.wav` file with the same sample rate.

### Usage:
Ensure that DeepFilterNet is installed and that you have a valid noisy audio file.
Modify the `noisy_audio_path` variable to point to your actual file before running the script.

- SoundFile (`soundfile`)
- PyTorch (`torch`)
- DeepFilterNet (`df` module)

### File Paths:

- Output: `enhanced_audio.wav` (Will be saved in the current directory)

   - Measures the perceived audio quality of the enhanced speech.
   - Score range: -0.5 to 4.5 (higher is better).
   - Commonly used in speech enhancement and VoIP applications.
   - `"wb"` (Wideband mode) is used for 16kHz sampling rate; `"nb"` (Narrowband) is for 8kHz.

2. **STOI (Short-Time Objective Intelligibility)**
   - Measures the speech intelligibility improvement.
   - Score range: 0 to 1 (higher is better).
   - A score above **0.75** indicates good intelligibility.

3. **SDR (Signal-to-Distortion Ratio)**
   - Measures how much the enhanced speech resembles the original.
   - Score is in decibels (dB), with **higher values indicating better enhancement**.
   - **10 dB or higher** is considered good, indicating reduced noise and distortion.

# Step 2: Batch Speech Enhancement and Evaluation using DeepFilterNet3

"""
Batch Speech Enhancement and Evaluation using DeepFilterNet3

This script performs batch speech enhancement on a folder of noisy audio files
using the DeepFilterNet3 model. Each enhanced output is saved to a specified directory,
and several speech quality/intelligibility metrics are computed to assess performance.

üìÅ Input:
- Directory of noisy `.wav` files (`NOISY_DIR`)
- Each file is assumed to be mono-channel and in a valid WAV format

üìÅ Output:
- Enhanced audio saved to `OUTPUT_DIR` with filenames prefixed as `enhanced_`
- A CSV file `enhancement_results.csv` summarizing evaluation metrics

üîç Enhancement Process:
1. Loads each noisy file using `soundfile` (preserving native sample rate)
2. Converts the audio to a PyTorch tensor and processes it with DeepFilterNet3
3. Saves the enhanced output using the original sample rate

üìä Evaluation Metrics (computed at 16 kHz):
- PESQ (Perceptual Evaluation of Speech Quality) - Wideband mode
- STOI (Short-Time Objective Intelligibility)
- SDR (Signal-to-Distortion Ratio) using mir_eval‚Äôs BSS_eval

üîß Requirements:
- Python packages: `torch`, `soundfile`, `librosa`, `numpy`, `pandas`, `df`, `pesq`, `pystoi`, `mir_eval`
- DeepFilterNet must be installed and accessible via `from df import init_df`

üìù Output Summary:
- A printed score for each file: PESQ, STOI, SDR
- A complete results table printed at the end and saved to `enhancement_results.csv`

üë§ Author: [Sahar Hashem]
üìÖ Date: [03/2025]
"""


````python
import os
import torch
import soundfile as sf
import librosa
import numpy as np
import pandas as pd
from df import init_df
from df.enhance import enhance
from pesq import pesq
from pystoi import stoi
from mir_eval.separation import bss_eval_sources

# Define input and output directories
NOISY_DIR = "/Users/Sahar/voice-enhance-mvp/noisy_testset_wav"  # Folder containing noisy .wav files
OUTPUT_DIR = "enhanced_audio_files"  # Folder to save enhanced .wav files
os.makedirs(OUTPUT_DIR, exist_ok=True)

# List to hold evaluation results
results = []

# Process each .wav file in the directory
for filename in os.listdir(NOISY_DIR):
    if filename.endswith(".wav"):
        noisy_path = os.path.join(NOISY_DIR, filename)
        enhanced_path = os.path.join(OUTPUT_DIR, f"enhanced_{filename}")

        print(f"\nüîπ Processing {filename}...")

        # Initialize DeepFilterNet3 model for each file (if your model is stateful)
        model, df_state, _ = init_df()

        # Load the noisy audio using SoundFile (to use the file's native sample rate)
        noisy_audio, sr_orig = sf.read(noisy_path)

        # Convert the audio to a tensor and add a batch dimension
        noisy_tensor = torch.tensor(noisy_audio, dtype=torch.float32).unsqueeze(0)

        # Enhance audio (wrapped in no_grad to avoid gradient computations)
        with torch.no_grad():
            enhanced_audio = enhance(model, df_state, noisy_tensor)
        enhanced_audio_np = enhanced_audio.squeeze(0).numpy()

        # Save the enhanced audio with the original sample rate
        sf.write(enhanced_path, enhanced_audio_np, sr_orig)
        print(f"‚úÖ Enhanced {filename} -> Saved to {enhanced_path}")

        # For evaluation, load both the noisy and enhanced audio at 16kHz (required by PESQ)
        noisy_eval, sr_eval = librosa.load(noisy_path, sr=16000)
        enhanced_eval, _ = librosa.load(enhanced_path, sr=16000)

        # Compute evaluation metrics
        pesq_score = pesq(sr_eval, noisy_eval, enhanced_eval, "wb")  # PESQ in Wideband mode
        stoi_score = stoi(noisy_eval, enhanced_eval, sr_eval, extended=False)
        sdr, _, _, _ = bss_eval_sources(
            np.expand_dims(noisy_eval, axis=0),
            np.expand_dims(enhanced_eval, axis=0)
        )

        # Append results for this file
        results.append([filename, pesq_score, stoi_score, sdr[0]])
        print(f"‚úÖ PESQ: {pesq_score:.3f}, STOI: {stoi_score:.3f}, SDR: {sdr[0]:.3f} dB")

# Save all results to a CSV file
df_results = pd.DataFrame(results, columns=["File", "PESQ", "STOI", "SDR"])
df_results.to_csv("enhancement_results.csv", index=False)

print("\nüîπ Batch Processing Complete!")
print(df_results)
````
